```python
%sql
CREATE OR REPLACE TABLE gold_sales_by_property AS
SELECT
    Property,
    COUNT(*) AS total_sales,
    AVG(Sale_Price) AS avg_price,
    AVG(carpet_area) AS avg_size
FROM property_transactions_silver
GROUP BY Property;
select * from gold_sales_by_property
limit 10;

CREATE OR REPLACE TABLE gold_sales_by_year AS
SELECT
    YEAR(Date) AS year,
    COUNT(*) AS total_sales,
    AVG(Sale_Price) AS avg_price
FROM property_transactions_silver
GROUP BY YEAR(Date)
ORDER BY year;
select * from gold_sales_by_year
limit 10;

CREATE OR REPLACE TABLE gold_price_by_rooms AS
SELECT
    num_rooms,
    AVG(Sale_Price) AS avg_price,
    COUNT(*) AS total_sales
FROM property_transactions_silver
GROUP BY num_rooms
ORDER BY num_rooms;
select * from gold_price_by_rooms
limit 10;
```

```python
%sql
CREATE OR REPLACE TABLE gold_sales_by_locality AS
SELECT
    Locality,
    COUNT(*) AS total_sales,
    AVG(Sale_Price) AS avg_sale_price,
    AVG(Estimated_Value) AS avg_estimated_value
FROM property_transactions_silver
WHERE Locality IS NOT NULL
GROUP BY Locality;

SELECT * FROM gold_sales_by_locality;
```
